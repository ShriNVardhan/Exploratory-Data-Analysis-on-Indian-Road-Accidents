# **Demographic Analysis**

Analysing the distribution of accidents by the 'Age_band_of_driver', 'Sex_of_driver', and 'Educational_level'. We will also be able to understand the correlation between the columns mentioned above.


```python
accidents['Age_band_of_casualty'].unique()
```

```python
accidents_age_order = ['Under 18', '18-30', '31-50','Over 51','Unknown']

plt.figure(figsize=(9,8))
sns.countplot(x= 'Age_band_of_driver', data = accidents, order = accidents_age_order, palette = 'Reds', width = 0.5)
sns.set_style('darkgrid')

plt.title('Age Band of Accidents')
plt.xlabel('Age Band of Drivers')
plt.ylabel('Count of Accidents')
matplotlib.rcParams['font.size'] = 8


plt.show()
```

these accidents.
* The startling aspect is that, given that you cannot obtain a driver's license until you are eighteen, a significant number of these incidents involve teenagers.


```python
plt.figure(figsize = (5,5))
colors = ['#CD533B', '#E1CA96', '#A5B5BF']
accidents['Sex_of_driver'].value_counts().plot.pie(autopct='%1.01f%%', startangle=10, colors = colors)

plt.title('Accidents by Sex of Driver')
plt.ylabel(' ')
matplotlib.rcParams['font.size'] = 8
plt.show()
```

```python
accidents.Educational_level.unique()
```

```python
plt.figure(figsize=(10, 6))
order = ['Illiterate', 'Writing & reading', 'Elementary school', 'Junior high school', 'High school', 'Above high school']
sns.set_style("darkgrid")
sns.countplot(data=accidents, x='Educational_level', order= order, palette = 'Reds', width = 0.5, linewidth = 1)
plt.title("Relationship between Educational Level and Number of Accidents")
plt.xlabel('Educational Level')
plt.ylabel('Count of Accidents')
plt.show()
```

As is evident from this example, one's educational attainment has a significant impact on many aspects of life outcomes. It should come as no surprise that having greater education will result in far fewer accidents. You will undoubtedly drive more cautiously and with less disregard for the risks involved in getting from one location to another.

# **Time and Day Analysis**:
---

* Exploring the distribution of accidents over time using the 'Time' column.
* Analyze trends in accidents over days of the week.




```python
accidents.Time.value_counts().sort_values(ascending = True)
```

```python
accidents[['Hour', 'Minute', 'Second']] = accidents['Time'].str.split(':', expand=True)
accidents['Hour of the Day'] = accidents['Hour'].astype(int)

plt.figure(figsize=(12, 7))
sns.set_style("darkgrid")
sns.countplot(x='Hour of the Day', data=accidents, width = 0.7, palette = 'Reds', linewidth = 1.2)

plt.title('Accidents over Hours of the day')
plt.xlabel('Hour of the Day')
plt.ylabel('Count of Accidents')
matplotlib.rcParams['font.size'] =  9

plt.show()
```

Not surprising that a lot of accidents happen in the morning and late in the evening when people are going and coming back from work.

```python
days_ordered = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
accidents.Day_of_week = pd.Categorical(accidents['Day_of_week'], categories=days_ordered, ordered=True)

plt.figure(figsize = (7,7))
sns.set_style("darkgrid")
sns.countplot(x = 'Day_of_week', data = accidents, palette = 'Reds', width = 0.5, linewidth = 1.2)
plt.title('Accidents by Day of Week')
plt.ylabel('Count of Accidents')
plt.xlabel('Day of Week')
matplotlib.rcParams['font.size'] =  9
plt.tight_layout()
plt.show()
```

# **Cause and Severity Analysis**:
---
Deep diving on what these causes can be and study the correlation with the severity of the accident. We will be using 'Light_conditions', 'Cause_of_accident', 'Weather_condition' and 'Accident_severity' columns.  





```python
plt.figure(figsize = (7,7))
sns.set_style("darkgrid")
sns.countplot(x = 'Light_conditions', data = accidents, order = accidents['Light_conditions'].value_counts().index, palette = 'Reds',
              width = 0.4, linewidth = 1.3)
plt.title('Relation between Light Conditions and Number of Accidents')
plt.xlabel('Light Conditions')
plt.ylabel('Count of Accidents')
matplotlib.rcParams['font.size'] = 8

plt.show()
```

Unprecedented numbers of accidents occur both during the day and at night (when there is enough lighting). When there is little to no visibility, people usually take extra precautions or choose not to drive at all.


```python
plt.figure(figsize = (10,8))
sns.set_style("darkgrid")
sns.countplot(y = 'Cause_of_accident', data = accidents, order = accidents['Cause_of_accident'].value_counts().index, palette = 'Reds', linewidth = 1.4 )
plt.title("Relationship between Cause of Accident and the Number of Accidents")
plt.xlabel('Count of Accidents')
plt.ylabel('Cause of Accident')
plt.show()
```

```python
weather_sev_corr = pd.crosstab(accidents['Weather_conditions'], accidents['Accident_severity'])
plt.figure(figsize = (12,8))
sns.heatmap(weather_sev_corr, annot = True, cmap = 'Reds', fmt = 'd', linewidth = 1)
plt.title('Correlation Between Weather Conditions and Accident Severity')
plt.xlabel('Accident Severity')
plt.ylabel('Weather Condtions')
plt.show()
```

Not particularly a misconception but definitely dismisses the claim of accidents happening more often in bad weather conditions. It definitely plays a big role but reckless drivers can get themselves in trouble in any weather and even in the most perfect driving conditions.  

```python
acc_filtered = accidents[accidents['Driving_experience'] != 'Unknown']

acc_filtered['Driving_experience'] = acc_filtered['Driving_experience'].str.extract(r'(\d+)').astype(float)

plt.figure(figsize=(10, 6))
sns.violinplot(data=acc_filtered, x='Driving_experience', y='Accident_severity', palette = 'deep')
plt.xlabel("Years of Driving Experience")
plt.ylabel("Accident Severity")
plt.title("Relationship between Years of Driving Experience and Accident Severity")
plt.show()
```

Median of all three plots lie in the 2 year mark. The density of data points in the plots decreases sharply as your experience increases.  

# **Casualty Analysis**:

Exploring the casualty severity distribution and the correlation between the age band of drivers and the count of accidents taking place.

```python
plt.figure(figsize=(5, 5))

colors = ['#CD533B', '#E1CA96', '#A5B5BF']
accidents['Casualty_severity'].value_counts().plot.pie(autopct='%1.1f%%', startangle=25, colors = colors)
plt.title('Casualty Severity Distribution')
plt.ylabel(' ')
plt.xticks(rotation = 'right')
plt.show()
```

```python
age_band_order = ['5', 'Under 18', '18-30', '31-50', 'Over 51', 'na']
plt.figure(figsize=(8, 8))
sns.set_style("darkgrid")
sns.countplot(x='Age_band_of_casualty', data=accidents, order = age_band_order, palette = 'Reds', linewidth = 1.3, width = 0.6)
plt.title('Casualties by Age Band')
plt.xlabel('Age band of driver')
plt.ylabel('Count of Accidents')
plt.show()
```

Big chunk of casualties in the 18-30 age band which is unfortunately not surprising at all.
